##4.เปรียบเทียบเเวอร์ชั่น (diff)
ใช้สำหรับครวจสอบและเปรียบเทียบไฟล์โค๊ดว่ามีอะไรเปลี่ยนแปลงและแตกต่างไปจากเดิมเมื่อเที่ยบกับ Commit ที่ผ่านมา
* `git diff <commit_id>` แบบระบุ commit ID
* `git diff <commit_id><commit_id>` เปรียบเทียบระหว่าง 2 commit

`--` คือ version เก่า

`++` คือ version ใหม่


```python
git diff 2e5b2b7 4b6923e
```

##7.Git Branching
Git Branching คือการสร้างสาขา (branch) เพื่อพัฒนาฟีเจอร์หรือทดลองโค้ดแยกจากสาขาหลัก (main/master) โดยแต่ละสาขาทำงานแยกกันได้อย่างอิสระ ช่วยลดปัญหาเมื่อต้องรวมโค้ดกลับ (merge) เข้าด้วยกันในภายหลัง

คำสั่งที่เกี่ยวข้อง

* แสดงชื่อ
`git branch `
1. สร้าง branch
`git branch ชื่อbranch`
`
2. เปลี่ยนไปทำงาน branch
`git checkout -b ชื่อbranch`

3. สลับไป branch หลัก
`git checkout master`

4. รวม branch
`git merage ชื่อbranch`

5. การลบ branch
`git branch -d ชื่อbranch`
